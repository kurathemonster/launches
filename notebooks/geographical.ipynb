### Geospatial

This notebook is for mapping out launch data based on physical locations on a map. Here I will correlate where launches are concentrated in the world and any patterns seen in the locations of launches.

```python
# Setup

import pandas as pd
import numpy as np
import folium

df = pd.read_csv('../data/space_devs_launches_cleaned.csv')
```

```python
df = df[['launch_name', 'launch_pad', 'launch_location', 'launch_country', 'pad_latitude', 'pad_longitude']]

map_df = df.dropna(subset=['pad_latitude', 'pad_longitude']).copy()

map_df
```

```python
# Map of all launch locations - most are on the coast or even in the ocean

map = folium.Map([10, -70], zoom_start=3)

for index, row in map_df.iterrows():
    folium.Marker([row['pad_latitude'], row['pad_longitude']], popup=row['launch_pad'], tooltip=row['launch_location']).add_to(map)
map
```

```python
# Heat map

from folium.plugins import HeatMap

heat_map = folium.Map(
    locations=[20, 0],
    zoom_start=2,
    titles="CastoDB positron"
)

# Create heatmap data
heat_data = map_df[['pad_latitude', 'pad_longitude']].values.tolist()

# Add heatmap layer
HeatMap(heat_data).add_to(heat_map)

heat_map

# Launches are concentrated in the eastern US and eastern Asia (Korea, Japan, China)
```

```python
# Looking at the top launch sites

top_sites = df['launch_location'].value_counts().head(10)
top_pads = df[['launch_pad', 'launch_location']].value_counts().head(10)

print(f"Top Sites\n{top_sites}\n")
print(f"Top Pads\n{top_pads}")
```
